```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\Lenovo\ \Trumpler 10.csv")
print (data.columns)
print (data.head())

absolute_magnitude=data['phot_g_mean_mag']
color_index = data['bp_rp']

plt.figure(figsize=(20,6))
plt.scatter (color_index, absolute_magnitude, s=50, c='blue', alpha=0.7)
plt.gca().invert_yaxis()
plt.title('Hertzsprung-Russell Diagram')
plt.xlabel('Color Index (B-V)')
plt.ylabel('Absolute Visual Magnitude')
plt.grid(True)
plt.savefig('Trumpler 10.png')
plt.show()
```
